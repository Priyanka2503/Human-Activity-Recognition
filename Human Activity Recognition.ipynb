```python
# Importing useful packages
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
#Loading of dataset
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
print(train.shape, test.shape)
```

```python
#Checking of Null Values
print("Null values present in training data: {}".format(train.isnull().values.any()))
print("Null values present in testing data: {}".format(test.isnull().values.any()))
```

```python
train.head()
```

```python
test.head()
```

```python
#Getting training data 
trainData  = train.drop(['Activity','ActivityName'] , axis=1).values
trainLabel = train.Activity.values

#Getting test data
testData  = test.drop(['Activity','ActivityName'] , axis=1).values
testLabel = test.Activity.values
```

```python
#Random Forest Classifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200,  n_jobs=4, min_samples_leaf=10)    
#train
rf.fit(trainData, trainLabel)
#test


y_te_pred = rf.predict(testData)
acc = accuracy_score(testLabel, y_te_pred)

print("Accuracy: %3.5f" % (acc))
```

```python
# Support Vector Classifier
from sklearn.svm import SVC
svc = SVC() 
svc.fit(trainData, trainLabel)
y_pred = svc.predict(testData)
print('Model accuracy score:{}'. format(accuracy_score(testLabel, y_pred)))
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()

model.fit(trainData, trainLabel)
yhat=model.predict(testData)

lr_accuracy = accuracy_score(y_true=testLabel, y_pred=yhat)
print("Accuracy using Logistic Regression : ", lr_accuracy)
```

```python
#K Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier().fit(trainData, trainLabel)
prediction = clf.predict(testData)

print('K Nearest Neighbors Classifier accuracy: {}'.format(accuracy_score(testLabel, prediction)))
```
